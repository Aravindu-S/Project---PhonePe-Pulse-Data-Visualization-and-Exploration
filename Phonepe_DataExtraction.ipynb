## Aggregated Transaction

```python
# Aggregated Transaction

path1= "D:/Data science/Project/PhonePe/pulse/data/aggregated/transaction/country/india/state/"
agg_trans_list = os.listdir(path1)

table_1= {'States':[], 'Year':[], 'Quarter':[],"Transaction_name":[],"Transaction_count":[], 'Transaction_amount': []}

for states in agg_trans_list:
    current_state = path1+states+'/'

    #os.listdir() method in python is used to get the list of all files and directories in the specified directory.
    
    year_list = os.listdir(current_state)

    for year in year_list:
        current_year = current_state+year+'/'
        agg_trans_file_list = os.listdir(current_year)

        for file in agg_trans_file_list:
            current_file= current_year+file
            current_file_data=open(current_file)
            
            data1=json.load(current_file_data)

            for i in data1['data']['transactionData']:
                name=i['name']
                count=i['paymentInstruments'][0]['count']
                amount=i['paymentInstruments'][0]['amount']
                table_1['Transaction_name'].append(name)
                table_1['Transaction_count'].append(count)
                table_1['Transaction_amount'].append(amount)
                table_1['States'].append(states)
                table_1['Year'].append(year)
                table_1['Quarter'].append(file.strip('.json'))

aggre_trans_data=pd.DataFrame(table_1)


aggre_trans_data["States"] = aggre_trans_data["States"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
aggre_trans_data["States"] = aggre_trans_data["States"].str.replace("-"," ")
aggre_trans_data["States"] = aggre_trans_data["States"].str.title()
aggre_trans_data['States'] = aggre_trans_data['States'].str.replace("Dadra & Nagar Haveli & Daman & Diu","Dadra and Nagar Haveli and Daman and Diu")

# Creating Table in SQL

drop_query= '''DROP TABLE IF EXISTS aggre_trans_table'''
cursor.execute(drop_query)
db_1.commit()



query= '''CREATE TABLE IF NOT EXISTS aggre_trans_table(States VARCHAR(100),
                                                      Year INT, 
                                                      Quarter  INT,
                                                      Transaction_name VARCHAR(100),
                                                      Transaction_count bigint,
                                                      Transaction_amount bigint
                                                        )'''

cursor.execute(query)
db_1.commit()


for index,row in aggre_trans_data.iterrows():

    query= '''INSERT INTO aggre_trans_table (States,
                                            Year, 
                                            Quarter,
                                            Transaction_name,
                                            Transaction_count,
                                            Transaction_amount 
                                            )
                                            
                                            VALUES(%s,%s,%s,%s,%s,%s)''' 
    value=(row['States'],
           row['Year'],
           row['Quarter'],
           row['Transaction_name'],
           row['Transaction_count'],
           row['Transaction_amount']
           )
    
    cursor.execute(query,value)
    db_1.commit()
```

## Map User

```python
# Map User

path4= "D:/Data science/Project/PhonePe/pulse/data/map/user/hover/country/india/state/"

map_user_list = os.listdir(path4)

table_4= {'States':[], 'Year':[], 'Quarter':[],"District":[],"Registered_Users":[],'App_Opens': []}

for states in map_user_list:
    current_state = path4+states+'/'
    year_list = os.listdir(current_state)

    for year in year_list:
        current_year= current_state+year+'/'
        map_user_file_list = os.listdir(current_year)

        for file in map_user_file_list:
            current_file = current_year+file
            current_file_data=open(current_file)

            data4= json.load(current_file_data)

            for i in data4['data']['hoverData'].items():
                district=i[0]
                registered_Users= i[1]['registeredUsers']
                app_Opens= i[1]['appOpens']
                table_4['States'].append(states)
                table_4['Year'].append(year)
                table_4['Quarter'].append(file.strip('.json'))
                table_4['District'].append(district)
                table_4['Registered_Users'].append(registered_Users)
                table_4['App_Opens'].append(app_Opens)
                

        
map_user_data = pd.DataFrame(table_4)

map_user_data["States"] = map_user_data["States"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
map_user_data["States"] = map_user_data["States"].str.replace("-"," ")
map_user_data["States"] = map_user_data["States"].str.title()
map_user_data['States'] = map_user_data['States'].str.replace("Dadra & Nagar Haveli & Daman & Diu","Dadra and Nagar Haveli and Daman and Diu")

# Creating table in SQL
                
drop_query = '''DROP TABLE IF EXISTS map_user_table'''

cursor.execute(drop_query)
db_1.commit()


query = ''' CREATE TABLE IF NOT EXISTS map_user_table(States  VARCHAR(50),
                                                      Year    INT,
                                                      Quarter INT,
                                                      District VARCHAR(100),
                                                      Registered_Users BIGINT,
                                                      App_Opens BIGINT
                                                      )'''
cursor.execute(query)
db_1.commit()


for index,row in map_user_data.iterrows():

    query = '''INSERT INTO map_user_table(States,
                                          Year,
                                          Quarter,
                                          District,
                                          Registered_Users,
                                          App_Opens)
                                          
                                          VALUES(%s,%s,%s,%s,%s,%s)'''
    
    value= (row['States'],row['Year'],row['Quarter'],row['District'],row['Registered_Users'],row['App_Opens'])

    cursor.execute(query,value)
    db_1.commit()

```
